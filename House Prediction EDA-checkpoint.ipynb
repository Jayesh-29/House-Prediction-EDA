```python
discrete_features = [i for i in numerical_features if len(df[i].unique()) < 20]
print(f"Total discrete feature are: {len(discrete_features)}")
print(discrete_features)
```

```python
continuous_features = [i for i in numerical_features if i not in discrete_features ]
print(f"Total continuous feature are: {len(continuous_features)}")
print(continuous_features)
```

```python
def discrete_analysis(df,i):
    plt.figure(figsize=(10,5))
    df.groupby(i)['SalePrice'].median().plot.bar()
    plt.ylabel("Sale Price")
    plt.xlabel(i)
    plt.title('Sale Price according to '+ i)
    plt.show()
for i in discrete_features:
    discrete_analysis(train,i)
```

```python
def continuous_analysis(df,i):
    plt.figure(figsize=(10,5))
    df[i].hist()
    plt.ylabel("Number of Houses")
    plt.xlabel(i)
    plt.show()
for i in continuous_features:
    continuous_analysis(train,i)
```
